```python
data = {
    'Year': [2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012],
    'Total': [0,    1,    4,   10,  150,  250,  350,  450,  550,  650,  750,  850]
   
}

#Convert data to Dataframe
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Total'], marker='o', color='b', linestyle='-')
plt.title('Total Suicides Over the Years')
plt.xlabel('Year')
plt.ylabel('Total')
plt.grid(True)
plt.xticks(df['Year'])
plt.tight_layout()
plt.show()
```

```python
data = {
    'State': ['A & N ISLANDS', 'A & N ISLANDS', 'A & N ISLANDS', 'A & N ISLANDS', 'A & N ISLANDS', 'TAMIL NADU', 'TAMIL NADU', 'TAMIL NADU', 'TAMIL NADU', 'TAMIL NADU'],
    'Type': ['Cancer', 'Divorce', 'Dowry dispute', 'Ideological Causes/Hero Worshipping', 'Illness (Aids/STD)', 'Cancer', 'Divorce', 'Dowry dispute', 'Ideological Causes/Hero Worshipping', 'Illness (Aids/STD)'],
    'Number': [10, 5, 15, 8, 7, 20, 18, 22, 10, 12]
}

df = pd.DataFrame(data)
df.set_index('State', inplace=True)

fig, ax = plt.subplots(figsize=(10, 6))

df.plot(kind='bar', stacked=True, ax=ax)

plt.xlabel('State')
plt.ylabel('Number of Suicides')
plt.title('Distribution of Suicide Types Across Different States')
plt.legend(title='Suicide Type')

plt.show()
```

```python
states = ['A & N Islands', 'Bihar', 'Chhattisgarh', 'Delhi', 'Goa']
male_suicides = [29, 1537, 1958, 1177, 107]
female_suicides = [60, 1085, 1005, 355, 16]

# Calculate the male-to-female suicide ratio
ratio = [m / f for m, f in zip(male_suicides, female_suicides)]

# Create a bar chart
plt.bar(states, ratio, color='blue')
plt.xlabel('Regions')
plt.ylabel('Male-to-Female Suicide Ratio')
plt.title('Male-to-Female Suicide Ratio by Region')
plt.show()
```

```python
# Sample data 
data = {
    'Age Group': ['0-14', '15-24', '25-34', '35-44', '45-54', '55-64', '65+'],
   'Type1': ['Cancer', 'Divorce', 'Dowry dispute', 'Ideological Causes/Hero Worshipping', 'Illness (Aids/STD)', 'Cancer', 'Divorce', 'Dowry dispute', 'Ideological Causes/Hero Worshipping', 'Illness (Aids/STD)'],
   'Type2': ['Cancer', 'Divorce', 'Dowry dispute', 'Ideological Causes/Hero Worshipping', 'Illness (Aids/STD)', 'Cancer', 'Divorce', 'Dowry dispute', 'Ideological Causes/Hero Worshipping', 'Illness (Aids/STD)'],
   'Type3': ['Cancer', 'Divorce', 'Dowry dispute', 'Ideological Causes/Hero Worshipping', 'Illness (Aids/STD)', 'Cancer', 'Divorce', 'Dowry dispute', 'Ideological Causes/Hero Worshipping', 'Illness (Aids/STD)'],

       
}

df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")

# Use countplot to show distribution of suicide types in each age group
sns.countplot(x='Age Group', y='Type1', data=df, color='blue', label='Type 1')
sns.countplot(x='Age Group', y='Type2', data=df, color='orange', bottom=df['Type1'], label='Type 2')
sns.countplot(x='Age Group', y='Type3', data=df, color='green', bottom=df['Type1'] + df['Type2'], label='Type 3')

plt.title('Distribution of Suicide Types Among Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Number of Suicides')
plt.legend()

plt.show()
```

```python
# 'Type_code' is the column representing type codes
type_code_counts = df['Type_code'].value_counts()

# Bar Chart
plt.figure(figsize=(10, 6))
type_code_counts.plot(kind='bar')
plt.title('Frequency of Type Codes')
plt.xlabel('Type Codes')
plt.ylabel('Count')
plt.show()
```

```python
# Scatter plot with regression line
sns.regplot(x='Causes Type Code', y='Suicide Rates', data=df)
plt.title('Correlation between Causes Type Code and Suicide Rates')
plt.xlabel('Causes Type Code')
plt.ylabel('Suicide Rates')
plt.show()
```
