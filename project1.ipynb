```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

df = pd.read_csv('final.csv')
df.head()
```

```python
#Question1
#sample data
total_manpower = 100
operational_ratio = 0.7
high_priority_ratio = 0.3

#calculate manpower distribution
operational_manpower = int(total_manpower * operational_ratio)
high_priority_manpower = int(total_manpower * high_priority_ratio)


#plotting the bar chart
labels = ['Operational', 'High-Priority']
manpower_values = [operational_manpower, high_priority_manpower]


plt.bar(labels, manpower_values, color=['blue', 'red'])
plt.xlabel('Role Type')
plt.ylabel('Planned Manpower')
plt.title('Distribution of Planned Manpower between Operational and High_Priority Roles')
for i, value in enumerate(manpower_values):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')

plt.show()
```

```python

import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime

#Question2

#sample data
dates = ['2021-02-01','2021-02-01','2021-02-01','2021-02-01','2021-02-01']
Actual_manpower = [3,2,2,3,3]

# convert dates to datetime objects
dates = [datetime.strptime(date, '%Y-%m-%d') for date in dates]

#Create a dataframe
data = pd.DataFrame({'Start Date': dates, 'Actual Manpower': Actual_manpower})

# Scatter plot for start date vs actual manpower
plt.figure(figsize=(10, 6))
plt.scatter(data['Start Date'], data['Actual Manpower'], marker='o')

plt.xlabel('Start Date')
plt.ylabel('Actual Manpower')
plt.title('Correlation between Start Date and ACtual Manpower for High-Priority Roles')
plt.grid(True)
plt.show()

```

```python
#Question 3
#sample data
data = {
    'Production Line': ['L-01','L-02','L-02','L-03','L-04'],
    'Accumulated Hour': [8.617256,9.704712,1.831078,9.397757,1.257576],
    'Target Hour': [15.694828,4.709156,0.000000,17.545858,16.977273],
    'Working-Hour': [8,8,8,8,8],
    'Floor': ['Unit: 01(Mr.Liton)','Unit: 01(Mr.Liton)','Unit: 01(Mr.Liton)','Unit: 01(Mr.Liton)','Unit: 01(Mr.Liton)'],
}

df = pd.DataFrame(data)

#Trend in accumulated hours over different production lines(bar plot)
plt.figure(figsize=(10, 6))
sns.barplot(x='Production Line', y='Accumulated Hours', data=df, palette='viridis')
plt.title('Trend in Accumulated Hours over Different Production Lines')
plt.xlabel('Production Line')
plt.ylabel('Accumulated Hours')
plt.show()

#Varianace between target hours and actual working hours (barplot)
df['variance'] = df['ActualWorkingHours'] - df['TargetHours']
plt.figure(figsize=(10, 6))
sns.barplot(x='ProductionLine', y='Variance', data=df, palette='coolwarm')
plt.title('Variance between Target Hours and Actual Working Hours')
plt.xlabel('Production Line')
plt.ylabel('Variance (Actual - Target)')
plt.show()

#workforce distribution across different floors (pie chart)
floor_distribution = df['Floor'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(floor_distribution, labels=floor_distribution.index, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Workforce Distribution across Different Floors')
plt.show()
```

```python

import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime

#Question7
# Sample data


dates = ['2021-02-01','2021-02-01','2021-02-01','2021-02-01','2021-02-01']
Target_hours = [15.694828,4.709156,0.000000,17.545858,16.977273]
Actual_manpower = [3,2,2,3,3]

# convert dates to datetime objects
dates = [datetime.strptime(date, '%Y-%m-%d') for date in dates]

#Create a dataframe
data = pd.DataFrame({'Date': dates, 'Target Hours': Target_hours, 'Actual Manpower': Actual_manpower})
data = data.set_index('Date')

#Line chart for actual manpower vs target hours over time
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Target Hours'], label='Target Hours', marker='o')
plt.plot(data.index, data['Actual Manpower'], label='Actual Manpower', marker='o')


plt.xlabel('Date')
plt.ylabel('Hours')
plt.title('Actual Manpower vs Target Hours Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
